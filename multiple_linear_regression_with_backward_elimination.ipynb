```python
# importing libraries  
import numpy as np
import matplotlib.pyplot as mtplt 
import pandas as pd  
```

```python
#importing datasets  
data_set= pd.read_csv('D:\Project\MLPractice\dataset/50_Startups.csv')  
```

```python
#Extracting Independent and dependent Variable  
x= data_set.iloc[:, :-1].values  
y= data_set.iloc[:, 4].values 
```

```python
#Managing the Catgorical data  
#from sklearn.preprocessing import LabelEncoder, OneHotEncoder  
#labelencoder_x= LabelEncoder()  
#x[:, 3]= labelencoder_x.fit_transform(x[:,3])  
#onehotencoder= OneHotEncoder(categorical_features= [3])    
#x= onehotencoder.fit_transform(x).toarray()  
```

```python
#Managing the Catgorical data  
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

labelencoder_x= LabelEncoder()
x[:, 3]= labelencoder_x.fit_transform(x[:,3])
transformer = ColumnTransformer(
    transformers=[
        ("State",        # State
         OneHotEncoder(), # The transformer class
         [3]              # The column(s) to be applied on.
         )
    ],
    remainder='passthrough'
)
x = transformer.fit_transform(x.tolist())
x = x.astype('int64')
```

```python
#avoiding the dummy variable trap:  
x = x[:, 1:]  
```

```python
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state=0)  
```

```python
#Fitting the MLR model to the training set:  
from sklearn.linear_model import LinearRegression  
regressor= LinearRegression()  
regressor.fit(x_train, y_train)
```

```python
#Predicting the Test set result;  
y_pred= regressor.predict(x_test)  
```

```python
print('Train Score: ', regressor.score(x_train, y_train))  
print('Test Score: ', regressor.score(x_test, y_test))  
```

```python
# importing libraries  
import numpy as nm  
import matplotlib.pyplot as mtp  
import pandas as pd  
  
#importing datasets  
data_set= pd.read_csv('D:\Project\MLPractice\dataset/50_Startups.csv')   
  
#Extracting Independent and dependent Variable  
x_BE= data_set.iloc[:, 0:1].values  
y_BE= data_set.iloc[:, -1].values  
  
  
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_BE_train, x_BE_test, y_BE_train, y_BE_test= train_test_split(x_BE, y_BE, test_size= 0.2, random_state=0)  
  
#Fitting the MLR model to the training set:  
from sklearn.linear_model import LinearRegression  
regressor= LinearRegression()  
regressor.fit(nm.array(x_BE_train).reshape(-1,1), y_BE_train)  
  
#Predicting the Test set result;  
y_pred= regressor.predict(x_BE_test)  
  
#Cheking the score  
print('Train Score: ', regressor.score(x_BE_train, y_BE_train))  
print('Test Score: ', regressor.score(x_BE_test, y_BE_test)) 
```
